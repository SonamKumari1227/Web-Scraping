```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

webpage=requests.get('https://www.holidify.com/country/india/places-to-visit.html',headers=headers).text;
webpage
```

```python
soup=BeautifulSoup(webpage,'lxml')
soup
```

```python
print(soup.prettify())
```

```python
soup.find_all('card-heading')
```

```python
count=1
place=[]
for i in soup.find_all('h3',class_='card-heading'):
    if(count<10):
        place.append(i.text.strip()[2:])
    else:
         place.append(i.text.strip()[4:])
    count+=1
```

```python
place
```

```python
rank=[]
for i in range(1,43):
    rank.append(i)
rank
```

```python
soup.find_all('p',class_='mb-3')[0].text.strip()[12:]
```

```python
best_time=[]
for i in soup.find_all('p',class_='mb-3'):
    best_time.append(i.text.strip()[12:])
best_time
```

```python
soup.find_all('img')
```

```python
soup.find_all('img',class_='card-img-top lazy')[0]['data-original']
```

```python
Img_Url=[]
for i in soup.find_all('img',class_='card-img-top lazy'):
       Img_Url.append(i['data-original'])
Img_Url
```

```python
data={'place':place,'Rank':rank,'Best_Time':best_time,'Image_Url':Img_Url}
```

```python
df=pd.DataFrame(data)
df
```
